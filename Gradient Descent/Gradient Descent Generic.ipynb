```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
data = load_boston()
```

```python
X = data.data
Y = data.target
```

```python
X.shape, Y.shape
```

```python
#adding a column of 1s to replace the parameter c
X = np.append(X, np.ones(X.shape[0]).reshape(-1, 1), axis = 1)
```

```python
def cost(X, Y, m):
    total_cost = ((Y - (m*X).sum(axis = 1))**2).mean()
    return total_cost
    
```

```python
def step_gradient_descent(X, Y, m, learning_rate):
    N = len(X)    
    m_grad = np.zeros(X.shape[1])
    for j in range(len(m_grad)):
        for i in range(N):
            m_grad[j] += ((Y[i] - (m*X[i]).sum())*X[i][j])
    m = m - learning_rate*(-2/N)*m_grad
    print(cost(X, Y, m))
    return m
```

```python
def gradient_descent(X, Y, learning_rate, num_iterations):
    m = np.zeros(X.shape[1])
    for i in range(num_iterations):
        m = step_gradient_descent(X, Y, m, learning_rate)
    return m
```

```python
def run():
    learning_rate = 0.000003
    num_iterations = 1000
    m = gradient_descent(X, Y, learning_rate, num_iterations)
    return m
```

```python
m = run()
```
