```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(0, 100)
Y = np.arange(0, 100) + 5*np.random.randn(100)
```

```python
plt.plot(X, Y)
plt.show()
```

```python
def cost(X, Y, m, c):
    total_cost = ((Y - m*X)**2).sum()
    return total_cost
    
```

```python
def step_gradient_descent(X, Y, m, c, learning_rate):
    N = len(X)
    m_grad = (Y - m*X - c)*(X)
    c_grad = ((Y - m*X - c))
    m = m - (learning_rate)*(-2/N)*(m_grad.sum())
    c = c - (learning_rate)*(-2/N)*(c_grad.sum())
    print(cost(X, Y, m, c))
    return m, c
```

```python
def gradient_descent(X, Y, learning_rate, num_iterations):
    m = 0
    c = 0
    for i in range(num_iterations):
        m, c = step_gradient_descent(X, Y, m, c, learning_rate)
    return m, c
```

```python
def run():
    learning_rate = 0.0001
    num_iterations = 1000
    m, c = gradient_descent(X, Y, learning_rate, num_iterations)
    return m, c
```

```python
m, c = run()
```
